```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import re
import pandas as pd
import numpy as np
```

```python
# Reading the url
HttpObject= urlopen("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
movies_data=HttpObject.read()
print(movies_data)
```

```python
soup_data=soup(movies_data,'html.parser')
print(soup_data)
```

```python
containers = soup_data.findAll('div',{'class':'lister-item mode-advanced'})
print(type(containers),len(containers))
```

```python
containers[0]
```

```python
for contain in containers:
    movies_title = contain.find('h3',{'class':'lister-item-header'})
    Movies_Title=contain.h3.a.text
    Releasing_Year=contain.h3.find('span',{'class':'lister-item-year text-muted unbold'}).text.replace('(','').replace(')','')
    Rating=contain.find('div',{'class':'inline-block ratings-imdb-rating'}).text.strip()
    dir=contain.find('p',{'class':''}).a.text
    cast=contain.findAll('p',{'class':''})[0].text.split('|')[1].split(':')[1]
    #print(''.join(cast))
    votes=contain.findAll('p',{'class':'sort-num_votes-visible'})[0].text.split()[1].replace(',','')
    gross=contain.findAll('p',{'class':'sort-num_votes-visible'})[0].text.split()[4]
    print('\n')
    
    
    print(Movies_Title,Releasing_Year,Rating,dir,cast,votes,gross)
    #print(movies_title.text.split('.')[1].strip())
```

```python
f=open('IMDB_Top_100_Movies.csv','wb')
f.write('Movies_Title,Releasing_Year,Ratings,Director,Cast,Votes,Gross\n'.encode())
for contain in containers:
    movies_title = contain.find('h3',{'class':'lister-item-header'})
    Movies_Title=contain.h3.a.text
    Releasing_Year=contain.h3.find('span',{'class':'lister-item-year text-muted unbold'}).text.replace('(','').replace(')','')
    Ratings=contain.find('div',{'class':'inline-block ratings-imdb-rating'}).text.strip()
    Director=contain.find('p',{'class':''}).a.text
    cast=contain.findAll('p',{'class':''})[0].text.split('|')[1].split(':')[1].split()
    Cast=''.join(cast)
    Votes=contain.findAll('p',{'class':'sort-num_votes-visible'})[0].text.split()[1].replace(',','')
    Gross=contain.findAll('p',{'class':'sort-num_votes-visible'})[0].text.split()[4]
    
    print(Movies_Title,Releasing_Year,Ratings,Director,Cast,Votes,Gross)
    f.write(f"{Movies_Title},{Releasing_Year},{Ratings},{Director},{Cast},{Votes},{Gross}\n".encode())
    print('\n')
f.close()
    
```
